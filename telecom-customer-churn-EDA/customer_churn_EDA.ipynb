Creating Boxplot for finding out the relation between Customer Churn(boolean data) and total charge rate(continuous data)

```python
# Box plot
plt.figure(figsize=(8,6))

sns.boxplot(x='Churn', y='Total Charge', data=df)
plt.title('Box Plot: Churn vs. Total Call Charge')
plt.show()
```

The box plot helps show the central tendency (median) and variability (IQR) in total call charges, along with any exceptional cases (outliers), giving us a clearer picture of how total call charge patterns differ between churned and non-churned customers.

**Higher Median for Churned Customers**: This suggests that customers with higher total call charges are more likely to churn.

**Greater Spread for Churned Customers**: The wider IQR indicates more variability in total call charges among churned customers, suggesting that high and unpredictable usage could be associated with churn.

**Outliers in Both Groups**: While both groups have outliers, churned customers tend to have more extreme high total call charges, which may be a factor in their decision to leave.

```python
# Violin plot
plt.figure(figsize=(8,6))

sns.violinplot(x='Churn', y='Total Charge', data=df)
plt.title('Violin Plot: Churn vs. Total Call Charge')
plt.show()
```

The violin plot, by visualizing density, gives a deeper view of how data is distributed within each churn group, beyond the basic median and quartiles shown in the box plot.

**Spread for churned customers (1)**: A broader range of total call charges, meaning more variability in their usage.

**Concentration for non-churned customers (0)**: Most non-churned customers have lower total call charges, suggesting a more stable or predictable usage pattern among those who stay.

Now, let's shift our focus to another key factor  **`Customer service calls`**. We'll create a summary table and a visual representation to better understand its impact on churn.

```python
sns.countplot(x="Customer service calls", hue="Churn", data=df);
```

```python
df["Higher service calls"]=(df["Customer service calls"]>3).astype("int64")
```

The code creates a new column, **`Higher service calls`**, where customers with more than 3 service calls are marked as `1`, and those with 3 or fewer calls are marked as `0`. It converts the boolean condition into integers for easier analysis.

```python
sns.countplot(x="Higher service calls", hue="Churn", data=df);
```

The **`pd.crosstab()`** function creates a table showing the frequency of churn based on whether customers made more than 3 service calls or not. The **`margins=True`** option adds totals for each row and column, providing an overall count for each category.

```python
pd.crosstab(df["Higher service calls"], df["Churn"], margins=True)
```

```python
pd.crosstab(df["Higher service calls"] & df["International plan"], df["Churn"])
```

So fromt the analysis above, we can say a customer is likely to leave (Churn = 1) based on two factors:

1. If the customer has made more than 3 calls to customer service and
2. If they are subscribed to an International Plan.

  If both of these are true, the prediction would be that the customer is not loyal and will leave (Churn = 1). If not, we predict that the customer will stay (Churn = 0).

Using this simple rule, the expected accuracy of the prediction is 85.75%, meaning that this method would be correct most of the time (except for a small number of cases, where we were wrong (373 + 7) = 380 times).This percentage is a good starting point to compare with when building more advanced machine learning models later.
